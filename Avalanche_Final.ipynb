```python
import numpy as np
from itertools import product

def int_to_bits(x, k):
    return np.array([int(b) for b in format(x, f'0{k}b')])

def bits_to_int(bits):
    return int(''.join(str(b) for b in bits), 2)

def linear_sub_cipher(x, a, b, N):
    return (a * x + b) % N

def avalanche_analysis(k):
    N = 2 ** k
    results = []
    possible_a = [a for a in range(3, N, 2)]  # Odd a, 3 <= a <= N-1
    possible_b = range(N)
    for a in possible_a:
        for b in possible_b:
            total_bit_flips = 0
            count = 0
            for x in range(N):
                x_bits = int_to_bits(x, k)
                y = linear_sub_cipher(x, a, b, N)
                y_bits = int_to_bits(y, k)
                for bit in range(k):
                    x_flipped = x ^ (1 << bit)
                    y_flipped = linear_sub_cipher(x_flipped, a, b, N)
                    y_flipped_bits = int_to_bits(y_flipped, k)
                    diff = np.sum(y_bits != y_flipped_bits)
                    total_bit_flips += diff
                    count += 1
            avg_flips = total_bit_flips / count
            results.append({'a': a, 'b': b, 'avg_flips': avg_flips})
    return results
# Run analysis for k=3
k3_results = avalanche_analysis(3)
for res in k3_results:
    print(f"a={res['a']}, b={res['b']}, Average Output Bits Flipped: {res['avg_flips']:.3f}")
```

```python
import random

def avalanche_analysis_k4(sample_keys=10):
    k = 4
    N = 2 ** k
    possible_a = [a for a in range(3, N, 2)]  # Odd a, 3 <= a <= N-1
    possible_b = range(N)
    all_keys = [(a, b) for a in possible_a for b in possible_b]
    # Randomly sample keys to keep computation reasonable
    sampled_keys = random.sample(all_keys, sample_keys)
    results = []
    for a, b in sampled_keys:
        total_bit_flips = 0
        count = 0
        for x in range(N):
            y = linear_sub_cipher(x, a, b, N)
            y_bits = int_to_bits(y, k)
            for bit in range(k):
                x_flipped = x ^ (1 << bit)
                y_flipped = linear_sub_cipher(x_flipped, a, b, N)
                y_flipped_bits = int_to_bits(y_flipped, k)
                diff = np.sum(y_bits != y_flipped_bits)
                total_bit_flips += diff
                count += 1
        avg_flips = total_bit_flips / count
        results.append({'a': a, 'b': b, 'avg_flips': avg_flips})
    return results

# Run analysis for k=4
k4_results = avalanche_analysis_k4(sample_keys=10)
for res in k4_results:
    print(f"a={res['a']}, b={res['b']}, Average Output Bits Flipped: {res['avg_flips']:.3f}")
```
